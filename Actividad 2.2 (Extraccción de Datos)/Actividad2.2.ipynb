```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb 
import numpy as np
```

```python
df= pd.read_excel('Datos de Facturación.xlsx')
```

```python
df.info()
```

```python
#CVE_CLPV (valores: 1000 a 2000) - Se tomaron los valores entre los 1000 y los 2000 de la columna CVE_CLPV
filtro1 = df[(pd.to_numeric(df["CVE_CLPV"], errors="coerce") >= 1000) & (pd.to_numeric(df["CVE_CLPV"], errors="coerce") <= 2000)]

filtro1.to_csv("Entregable1.csv")
filtro1
```

```python
#CVE_VEND (Todas las claves excepto “5” y “4”) - De la columna CVE_VEND se extrajeron las claves 5 y 4 para que no se mostraran en el reporte
filtro2=df[~df["CVE_VEND"].isin([5, 6])]

filtro2.to_csv("Entregable2.csv")
filtro2
```

```python
#FECHA_ENT (Fechas de “28/02/2022”) - De la columa FECHA_ENT solo se muestran los resultados de la fecha 28/02/2022
filtro3=df[(df["FECHA_ENT"] == "28/02/2022")]
filtro3.to_csv("Entregable3.csv")
filtro3
```

```python
#CAN_TOT (Cantidades menores a 5951.7) o STATUS (“E”) - - Para los valores menores a 595107 de la columna CAN_TOT o las filas que tengan como valor "E" en la columa STATUS
filtro4=df[(df["CAN_TOT"] > 5951.7) | (df["STATUS"] == "E")]

filtro4.to_csv("Entregable4.csv")
filtro4
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT - SE TOMAN CIERTAS COLUMNAS DEL DATASET PARA MOSTAR EL RESULTADO
filtro5=df.iloc[ : , [0, 6, 7, 9]]

filtro5.to_csv("Entregable5.csv")
filtro5
```

```python
#Solo las filas de 7001-7099 SOLO SE MUESTRAS LAS FILAS CONSECUTIVAS DE 7001 AL 7099
filtro6=df.iloc[7001:7100,:]

filtro6.to_csv("Entregable6.csv")
filtro6
```

```python
#Index= CVE_VEND (valores: 1, 2) (columna: FECHAELAB) 
filtro7 = df[df["CVE_VEND"].isin([1, 2])]
fechas_filtradas = filtro7["FECHAELAB"]

fechas_filtradas.to_csv("Entregable7.csv")
fechas_filtradas

```

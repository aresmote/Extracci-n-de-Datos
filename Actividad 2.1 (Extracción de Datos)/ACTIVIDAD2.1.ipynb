```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
df.info()
```

**FILTROS BÁSICOS PARA VARIABLES CUALITIVAS**

```python
#FILTRO POR OBJETOS PARA SABER EL NUMERO DE VENTAS DE UN SOLO VENDEDOR
filtro1=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]


#Convertir filtrado a archivo CSV
filtro1.to_csv("Entregable1.csv")

filtro1
```

```python
#FILTRO POR FILAS - Para que se muetren las columnas consecutivas de la 2 a la 8
#Filas Consecutivas 
filtro2= df.iloc[2:9,:]

#Filas Intermitentes
# filtro2= df.iloc[[3, 5, 7, 9],:]

filtro2.to_csv("Entregable2.csv")

filtro2
```

```python
#FILTRO POR COLUMNAS - Para que se muestran las columnas intermitententes 0, 2, 4, 6, 8
#Filas Continuas 
# filtro3=df.iloc[ : , 0:9]

#Filas Intermitentes
filtro3= df.iloc[ : , [0, 2, 4, 6, 8]]

filtro3.to_csv("Entregable3.csv")

filtro3
```

```python
#Copia del dataset comenzando por una columa etiquetada
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
df1
```

```python
#FILTRO 4 FILAS VS. COLUMNAS - Para mostrar ciertos valores de la columa NOMBRE_CLIENTE con la descripción de lo que están comprando
filtro4=df1.loc[["PÚBLICO EN GENERAL", "MOLINERA DE MEXICO", "INDUSTRIAS NIBA"], ["DESCR"]]

filtro4.to_csv("Entregable4.csv")

filtro4
```

```python
#FILTRO POR CABECERA - Para mostar las primeras filas del dataset
filtro5=df.head()

filtro5.to_csv("Entregable5.csv")

filtro5
```

```python
#FILTRO POR COLA - Para mostrar las ultimas filas del dataset
filtro6=df.tail()

filtro6.to_csv("Entregable6.csv")

filtro6
```

**FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS**

```python
#FILTRO POR COMPARACIÓN - Para las filas que tengan un valor menor o igual a 1000 en la columna COSTO_TOTAL_CALCULADO
filtro7=df[df["COSTO_TOTAL_CALCULADO"] <= 1000]

filtro7.to_csv("Entregable7.csv")

filtro7
```

**FILTROS CON CONECTORES LÓGICOS**

```python
#FILTRO Y - Para mostar las filas que en la columna CVE_ART tengan un valor igual a TUE00207 y en la columna COSTO_UNITARIO _CALCULADO tenga un valor $517.89
filtro8=df[(df["CVE_ART"] == "TUE00207") & (df["COSTO_UNITARIO_CALCULADO"] == 517.89)]

filtro8.to_csv("Entregable8.csv")

filtro8
```

```python
#FILTRO O - Para mostrar la columna CVE_ART con un valor igual a TUE00207 o que en la columa SUBTOTAL_PARTIDA tengan un valor mayor a 1000
filtro9=df[(df ["CVE_ART"] == "TUE00207") | (df["SUBTOTAL_PARTIDA"] > 1000)]

filtro9.to_csv("Entregable9.csv")

filtro9
```

```python
#FILTRO NOT - Para mostrar a los vendedores que no sean LETICIA RAMIREZ DURAN Y DIEGO ARMANDO LEON DURAN
filtro10=df[~df["NOMBRE_VENDEDOR"].isin(["LETICIA RAMIREZ HERNANDEZ", "DIEGO ARMANDO LEON DURAN"]) ]

filtro10.to_csv("Entregable10.csv")
filtro10
```

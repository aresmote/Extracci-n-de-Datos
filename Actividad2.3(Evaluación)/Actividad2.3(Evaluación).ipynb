**EXTRACCIÓN DE DATOS - OTTAWA CANADÁ** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Ottawa.csv')
```

```python
df.info()
```

**COMIENZO DE LOS FILTROS**


```python
#LOS REGISTROS DE 5 "HOST NAME" DIFERENTES
filtro1 = df['host_name'].dropna().unique()[:5] 
filtro1 = df[df['host_name'].isin(filtro1)]

filtro1.to_csv("Entregable1.csv")
filtro1
```

```python
#LOS REGISTROS DE LOS HOST QUE SE HALLAN UNIDO A AIRBNB "ANTES DEL AÑO 2019"
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce') 
filtro2 = df[df['host_since'] < '2019-01-01']

filtro2.to_csv("Entregable2.csv")
filtro2
```

```python
#LOS REGISTROS DE LOS HOST QUE RESPONDEN "A MÁS TARDAR EN 1 DÍA"
filtro3 = df[df['host_response_time'].str.contains('1 day', na=False, case=False)]

filtro3.to_csv("Entregable3.csv")
filtro3
```

```python
#LOS REGISTROS QUE TENGAN "AL MENOS 5 ACCOMMODATES"
filtro4 = df[df['accommodates'] >= 5]

filtro4.to_csv("Entregable4.csv")
filtro4
```

```python
#LOS REGISTROS DE LOS TIPOS DE CUARTO "ENTIRE HOME" Y "PRIVATE ROOM"
filtro5 = df[df['room_type'].isin(['Entire home/apt', 'Private room'])]

filtro5.to_csv("Entregable5.csv")
filtro5
```

```python
#LOS REGISTROS QUE "A LO MUCHO CUENTEN CON 4 CAMAS"
filtro6 = df[df['beds'] <= 4]

filtro6.to_csv("Entregable6.csv")
filtro6
```

```python
#LOS REGISTROS QUE "TENGAN 1 BAÑO O QUE SEAN SUPERHOST" PERO CON LA CONDICIÓN DE QUE "TODOS SEAN INSTANT_BOOKABLE"
filtro7 = ((df['bathrooms'] == 1) | (df['host_is_superhost'] == 't')) & (df['instant_bookable'] == 't')

filtro7.to_csv("Entregable7.csv")
filtro7
```

```python
#LOS REGISTROS CUYO "REVIEW_SCORES_CLEANLINESS SEA MENOR DE 3 Y REVIEW_SCORES_RATING SEA MAYOR DE 3"
filtro8 = (df['review_scores_cleanliness'] < 3) & (df['review_scores_rating'] > 3)

filtro8.to_csv("Entregable8.csv")
filtro8
```
